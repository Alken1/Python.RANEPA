```python
import numpy as np
st1=int(input('Введите первую сторону треугольника: '))
st2=int(input('Введите вторую  сторону треугольника: '))
st3=int(input('Введите третью сторону треугольника: '))
if st1< 0 or st2< 0 or st3< 0:
    while st1< 0 or st2< 0 or st3<0:
        print('Все стороны должны быть должны быть больше 0! Попробуйте снова!')
        st1=int(input('Введите первую сторону треугольника: '))
        st2=int(input('Введите вторую  сторону треугольника: '))
        st3=int(input('Введите третью сторону треугольника: '))
p_p=(int(st1)+int(st2)+int(st3))/2
s=np.sqrt(p_p*(p_p-int(st1))*(p_p-int(st2))*(p_p-int(st3)))
print('Площадь треугольника равна: ', s)
```

```python
#Напишите программу, которая находит все простые числа между 0 и числом, введенным пользователем.
n = input("n=")
n=int(n)
lst=[2]
for i in range(3, n+1, 2):
    if (i > 10) and (i%10==5):
        continue
    for j in lst:
        if j*j-1 > i:
            lst.append(i)
            break
        if (i % j == 0):
            break
    else:
        lst.append(i)
print (lst)
```

```python
#Напишите программу, которая спрашивает пользователя три числа (a, b, c), выдает решение квадратного уравнения. Или пишет, что решения нет.
a=input("Введите число a:")
b=input("Введите число b:")
c=input("Введите число c:")
a=int(a)
b=int(b)
c=int(c)
if a != 0:
    diskr=b*b-4*a*c
    if diskr>=0:
        print('x1=', (b +np.sqrt(diskr))/(2*a))
        print('x2=', (b -np.sqrt(diskr))/(2*a))
    else:
        print('Нет решений')
if a == 0 and b != 0:
    print('x=',-c/b)
if a == 0 and b == 0:
    print('Не уравнение')
```
